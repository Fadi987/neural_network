```python
from py_nn import Matrix, NeuralNetwork, MseOptimizer

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 100).reshape(-1, 1)

Y = 5*X**3 + 0.1 * np.random.randn(*X.shape)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(X, Y, 'ro')
plt.grid(True)
```

```python
X_matrix = Matrix.from_2d_vector(X.tolist())
Y_matrix = Matrix.from_2d_vector(Y.tolist())
```

```python
optimizer = MseOptimizer(0.01)

nn = NeuralNetwork()
nn.add_fully_connected_layer((10, 1))
nn.add_activation_layer('relu')
nn.add_fully_connected_layer((1, 10))
```

```python
for i in range(100):
    if i % 10 == 0:
        plt.figure(figsize=(20, 5))
        plt.plot(X, Y, 'ro')
        plt.plot(X, np.array(nn.predict(X_matrix).to_2d_list()).reshape(-1), label=f'Before Epoch {i}', linewidth=2)
        plt.grid(True)
        plt.legend()

    optimizer.train_on_sample(nn, X_matrix, Y_matrix)
```
